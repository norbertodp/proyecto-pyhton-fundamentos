```python
# Lista de diccionarios (cada diccionario es una fila)
empleados = [
    {"nombre": "Ana", "cargo": "Analista", "salario": 3500000},
    {"nombre": "Luis", "cargo": "Desarrollador", "salario": 4500000},
    {"nombre": "Sofía", "cargo": "Gerente", "salario": 6500000},
    {"nombre": "Carlos", "cargo": "Gerente I", "salario": 5500000},
    {"nombre": "Pedro", "cargo": "Gerente II", "salario": 4500000},
    {"nombre": "Mario", "cargo": "Gerente Junior", "salario": 3500000}
]

# Crear el DataFrame
df_empleados = pd.DataFrame(empleados)
df_empleados
```

```python
for  empleado in empleados:
    if empleado["cargo"]=="Desarrollador":
      print(empleado["nombre"])
      break
else:
  print("No se encontro gerente")
```

```python

found=False

for  empleado in empleados:
    if empleado["cargo"]=="Gerente":
      print(empleado["nombre"])
      found=True

if not found:
    print("No se encontro el empleado")

```

```python

# Buscar en un diccionario una coincidencia que no es exacta ejemplo solo encontro la palabra Gerenbte y trajo todas las busquedas parecidas

found=False

for empleado in empleados:
    if "Gerente" in empleado["cargo"]:
        print(empleado["nombre"])
        found=True

if not found:
    print("No se encontro el empleado")
```
